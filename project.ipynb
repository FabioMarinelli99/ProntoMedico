```python
import requests

def send_new_appointment():

  """Sends a PUT request to the specified URL with the appointment data as payload.

  Args:
    appointment: A dictionary containing the appointment data.
    url: The URL to send the request to.

  Returns:
    The server's response.
  """
  try:
    response = requests.put(server_url, json=appointment)
    response.raise_for_status()  # Raise an exception for bad status codes (4xx or 5xx)
    return response
  except requests.exceptions.RequestException as e:
    return f"Error: {response.status_code} - {response.text}"


def update_appointment(appointment_upd):

  """Sends a PUT request to the specified URL with the appointment data as payload.

  Args:
    appointment: A dictionary containing the appointment data.
    url: The URL to send the request to.

  Returns:
    The server's response.
  """
  try:
    response = requests.post(server_url, json=appointment_upd)
    response.raise_for_status()  # Raise an exception for bad status codes (4xx or 5xx)
    return response
  except requests.exceptions.RequestException as e:
    return f"Error: {response.status_code} - {response.text}"


def get_appointments():
    """
    Invia una richiesta GET al server per ottenere tutti gli appuntamenti.

    :param server_url: URL del server Flask (esempio: "http://127.0.0.1:5000")
    :return: La lista degli appuntamenti o un messaggio di errore
    """
    try:
        response = requests.get(f"{server_url}")
        if response.status_code == 200:
            return response.json()  # Restituisce i dati come un dizionario/JSON
        else:
            return {"error": f"Errore nella richiesta: {response.status_code}", "details": response.text}
    except requests.RequestException as e:
        return {"error": "Errore di connessione al server", "details": str(e)}
```

```python
# @title Parte del medico
import gradio as gr
from gradio_calendar import Calendar
from datetime import datetime, timedelta
import speech_recognition as sr

SAMPLE_REQUESTS = []

def get_urgency_indicator(urgency):
    if urgency >= 4:
        return "🔴"
    elif urgency >= 2:
        return "🟡"
    else:
        return "🟢"

def format_request_title(request):
    urgency_indicator = get_urgency_indicator(request["urgency"])
    received_datetime = datetime.strptime(request["received_date"], "%Y-%m-%d %H:%M:%S")
    received_info = received_datetime.strftime("%d/%m/%Y %H:%M")
    return f"{urgency_indicator} {request['patient_name']} (Urgenza: {request['urgency']}/5) - Ricevuto il: {received_info}"

def load_requests():
    global SAMPLE_REQUESTS
    # Ordina le richieste prima per urgenza (decrescente) e poi per data di ricezione (crescente)
    SAMPLE_REQUESTS = get_appointments()
    sorted_requests = sorted(
        SAMPLE_REQUESTS,
        key=lambda x: (-x["urgency"], x["received_date"])
    )
    for i, request in enumerate(sorted_requests):
        if request["status"] != "pending":
            sorted_requests.pop(i)
    titles = [format_request_title(req) for req in sorted_requests]
    return gr.update(choices=titles, value=titles[0] if titles else None)


def load_request_details(selected_title):
    if not selected_title:
        return "", "", gr.update(visible=False)

    for request in SAMPLE_REQUESTS:
        if request["patient_name"] in selected_title:
            details = f"""## 👤 {request['patient_name']}

**DESCRIZIONE:**
{request['description']}

**SUGGERIMENTO IA:**
{request['llm_suggestion']}

**DATA APPUNTAMENTO:** {request.get('appointment_date', 'Non specificata')}
**ORARIO:** {request.get('appointment_time', 'Non specificato')}"""
            return details, request['patient_name'], gr.update(visible=True)
    return "", "", gr.update(visible=False)

def remove_request(selected_title):
    if not selected_title:
        return gr.update(choices=[]), "", gr.update(visible=False), gr.update(visible=False), ""

    for i, request in enumerate(SAMPLE_REQUESTS):
        if request["patient_name"] in selected_title:
            update_appointment(SAMPLE_REQUESTS[i])
            break

    new_list = load_requests()

    return (
        new_list,
        "",
        gr.update(visible=False),
        gr.update(visible=False),
        "Richiesta chiusa con successo"
    )

def prepare_prescription(selected_title):
    if not selected_title:
        return gr.update(visible=False), "Seleziona prima una richiesta"
    return gr.update(visible=True), "Compila la ricetta per il paziente"

def confirm_appointment(selected_title):
    if not selected_title:
        return "Seleziona prima una richiesta"
    patient_name = next((req["patient_name"] for req in SAMPLE_REQUESTS if req["patient_name"] in selected_title), None)
    if patient_name:
        return f"Appuntamento confermato per {patient_name}. Il paziente riceverà una mail di conferma."

def call_patient(selected_title):
    if not selected_title:
        return "Seleziona prima una richiesta"
    patient_name = next((req["patient_name"] for req in SAMPLE_REQUESTS if req["patient_name"] in selected_title), None)
    if patient_name:
        return f"Chiamata urgente programmata per {patient_name}. Il paziente verrà contattato al più presto."

def generate_prescription(nome, indirizzo, citta, cap, provincia, codice_asl, farmaco, quantita):
    prescription = f"""SERVIZIO SANITARIO NAZIONALE               RICETTA ELETTRONICA
Regione Piemonte

COGNOME E NOME/INIZIALI DELL'ASSISTITO: {nome}
INDIRIZZO: {indirizzo}             CAP: {cap}        CITTÀ: {citta}        PROV: {provincia}
ESENZIONE:027           SIGLA PROVINCIALE: {provincia}      CODICE ASL: {codice_asl}

PRESCRIZIONE:
{farmaco}                                                  QTA: {quantita}

Rilasciato ai sensi dell'art.11, comma 16 del DL 31 mag 2010, n.78 e dell'art.1, comma 4 del DM 2 nov 2011
"""
    return "Ricetta generata con successo", prescription

with gr.Blocks(theme=gr.themes.Ocean()) as demo:
    gr.Markdown("# 👨‍⚕️ Pannello di Controllo Medico")

    with gr.Row():
        with gr.Column(scale=1):
            gr.Markdown("### 📋 Richieste in Attesa")
            requests_list = gr.Radio(
                choices=[],
                label="",
                interactive=True
            )
            refresh_button = gr.Button("🔄 Aggiorna Lista")

            gr.Markdown("""
            ### Legenda Urgenze
            🔴 Alta Urgenza (4-5)
            🟡 Media Urgenza (2-3)
            🟢 Bassa Urgenza (1)
            """)

        with gr.Column(scale=2):
            with gr.Row():
                gr.Markdown("### 📝 Dettagli Richiesta")
                close_button = gr.Button("Salva", scale=0)

            request_details = gr.Markdown()
            patient_name = gr.Textbox(visible=False)

            with gr.Row(visible=False) as action_buttons:
                prescription_btn = gr.Button("💊 Prepara Ricetta", variant="secondary")
                confirm_btn = gr.Button("✅ Conferma Appuntamento", variant="secondary")
                call_btn = gr.Button("📞 Chiama Paziente", variant="secondary")

            response_box = gr.Textbox(
                label="Stato Operazione",
                interactive=False,
                visible=True
            )

            # Sezione Ricetta (inizialmente nascosta)
            with gr.Column(visible=False) as prescription_section:
                gr.Markdown("### 📋 Compilazione Ricetta Elettronica")

                with gr.Row():
                    with gr.Column():
                        indirizzo = gr.Textbox(
                            label="Indirizzo",
                            placeholder="Via/Piazza..."
                        )
                        cap = gr.Textbox(
                            label="CAP",
                            placeholder="12345"
                        )
                        citta = gr.Textbox(
                            label="Città",
                            placeholder="Torino"
                        )

                    with gr.Column():
                        provincia = gr.Textbox(
                            label="Provincia",
                            placeholder="TO"
                        )
                        codice_asl = gr.Textbox(
                            label="Codice ASL",
                            placeholder="301"
                        )
                        farmaco = gr.Textbox(
                            label="Nome Farmaco",
                            placeholder="es. Augmentin 1g compresse"
                        )
                        quantita = gr.Number(
                            label="Quantità",
                            value=1,
                            minimum=1,
                            maximum=10,
                            step=1
                        )

                generate_btn = gr.Button("📄 Genera Ricetta", variant="primary")

                prescription_status = gr.Textbox(
                    label="Stato",
                    interactive=False,
                )

                prescription_preview = gr.Textbox(
                    label="Anteprima Ricetta",
                    interactive=False,
                    lines=10
                )

    # Eventi
    refresh_button.click(
        fn=load_requests,
        outputs=requests_list
    )

    requests_list.change(
        fn=load_request_details,
        inputs=requests_list,
        outputs=[request_details, patient_name, action_buttons]
    )

    prescription_btn.click(
        fn=prepare_prescription,
        inputs=[requests_list],
        outputs=[prescription_section, response_box]
    )

    confirm_btn.click(
        fn=confirm_appointment,
        inputs=[requests_list],
        outputs=response_box
    )

    call_btn.click(
        fn=call_patient,
        inputs=[requests_list],
        outputs=response_box
    )

    close_button.click(
        fn=remove_request,
        inputs=[requests_list],
        outputs=[
            requests_list,
            request_details,
            action_buttons,
            prescription_section,
            response_box
        ]
    )

    generate_btn.click(
        fn=generate_prescription,
        inputs=[
            patient_name,
            indirizzo,
            citta,
            cap,
            provincia,
            codice_asl,
            farmaco,
            quantita
        ],
        outputs=[prescription_status, prescription_preview]
    )

    # Carica le richieste all'avvio
    demo.load(fn=load_requests, outputs=requests_list)

demo.launch()
```

```python
# Sample available slots
AVAILABLE_SLOTS = {
    "2024-12-02": ["09:00", "10:00", "11:00", "15:00"],
    "2024-12-03": ["14:00 - 15:00", "15:00", "16:00"],
    "2024-12-04": ["09:00", "10:00", "11:00"],
    "2024-12-05": ["14:00", "15:00", "16:00"],
}

user_name = "Fabio Marinelli"
appointment_date = "none"
appointment_time = "none"
urgency = "none"
llm_suggestion = "none"
description = "none"

appointment = {
    "patient_name": user_name,
    "description": "none",
    "llm_suggestion": "none",
    "urgency": "none",
    "appointment_date": "none",
    "appointment_time": "none",
    "status": "pending"
}


@th.register_local_tool("send_appointment")
def send_new_appointment():

  """Sends a PUT request to the specified URL with the appointment data as payload.

  Args:
    appointment: A dictionary containing the appointment data.
    url: The URL to send the request to.

  Returns:
    The server's response.
  """
  try:
    response = requests.put(server_url, json=appointment)
    response.raise_for_status()  # Raise an exception for bad status codes (4xx or 5xx)
    return response
  except requests.exceptions.RequestException as e:
    return f"Error: {response.status_code} - {response.text}"

my_local_tools = [
    {
        "type": "function",
        "function": {
            "name": "send_appointment",
            "description": "Send the information about the appointment to the system so that the doctor can see the new appointment.",
            "parameters": {
                "type": "object",
                "properties": {
                    "required": [],
                },
            },
        },
    }
]

def get_summary(messages_story):
  system_prompt = """You are a chatbot designed to collect and process descriptions of physical
                    problems reported by patients of a general practitioner. Your task is to briefly
                    and effectively summarize the user's description, ensuring it is clear, concise,
                    and useful for the doctor. All responses must be written in Italian."""

  messages = [{"role": "system", "content": system_prompt}] + messages_story
  response = client.chat.completions.create(
      model=model,
      messages=messages,
      tools= th.get_tools(),
      temperature=0.7,
      max_tokens=256,
  )

  messages += th.run_tools(response)

  completion = client.chat.completions.create(
      model=model,
      messages=messages,
      tools= th.get_tools(),
      temperature=0.7,
      max_tokens=256,
  )

  response = completion.choices[0].message.content
  return response

summary_print = ""

general_error = ""

def get_medical_suggestion(summary):
    system_prompt = """You are a medical assistant that provides concise, actionable suggestions to doctors
                      based on technical summaries of patient cases, including diagnostic hypotheses, recommended
                      tests or treatments, and relevant considerations. All responses must be clear, professional,
                      concise, and presented in bullet points in Italian. Each suggestion should be a short phrase or a single word.
                      Additionally, provide an urgency level from 1 to 5, where 1 is low urgency and 5 is high urgency.
                      The response must be in the format: Consiglio: <here your advices>, Urgenza: <urgency>"""

    messages = [{"role": "system", "content": system_prompt}]
    messages.append({"role": "user", "content": summary})

    try:
        response = client.chat.completions.create(
            model=model,
            messages=messages,
            tools=th.get_tools() + my_local_tools,
            temperature=0.7,
            max_tokens=256,
        )

        messages += th.run_tools(response)

        completion = client.chat.completions.create(
            model=model,
            messages=messages,
            tools=th.get_tools() + my_local_tools,
            temperature=0.7,
            max_tokens=256,
        )

        response_content = completion.choices[0].message.content

        if "Consiglio:" in response_content and "Urgenza:" in response_content:
            parts = response_content.split("Urgenza:")
            suggestion = parts[0].replace("Consiglio:", "").strip()
            urgency = int(parts[1].strip())
        else:
            suggestion = response_content.strip()
            urgency = 1

        return [suggestion, urgency]
    except Exception as e:
        print(f"Error in get_medical_suggestion: {str(e)}")
        return ["Errore nel generare il suggerimento medico", 1]

class ChatHistory:
    def __init__(self):
        self.messages = []
        self.chat_display = []
        self.ai_message_count = 0
        self.ready_for_more_info = False
        self.ready_for_scheduling = False
        self.appointment_scheduled = False
        self.more_info_asked = False
        self.system_prompt = """You are a medical chatbot designed to collect detailed descriptions of health
        concerns from patients to send to their primary care physician. Your role is to engage with the patient
        in a natural, empathetic conversation.

        Follow these steps:
        1. Start by asking a natural question about their symptoms, without numbering or labeling the question.
        2. Listen to their response and ask a follow-up question related to what they shared, again keeping it
           natural and conversational.
        3. Based on their previous responses, ask one final question to gather any remaining important details.
        4. After they respond to all three questions, ask them "Vuoi aggiungere altre informazioni?"
        5. Regardless of their answer to the additional information question, proceed with scheduling by asking
           them to select an available date and time.
        6. After the appointment is scheduled, thank them for their time and confirm that their information
           will be sent to their doctor.

        Important guidelines:
        - All your responses must be in Italian
        - Never number your questions or indicate which question number you're on
        - Keep the conversation flowing naturally as a dialogue
        - Show empathy and understanding in your responses
        - Focus on gathering relevant medical information while maintaining a supportive tone"""

    def initialize(self):
        self.messages = [{"role": "system", "content": self.system_prompt}]
        self.chat_display = []
        self.ai_message_count = 0
        self.ready_for_more_info = False
        self.ready_for_scheduling = False
        self.appointment_scheduled = False
        self.more_info_asked = False

    def add_message(self, role, content):
        if not content:
            return

        self.messages.append({"role": role, "content": content})

        if role == "assistant":
            self.ai_message_count += 1
            if self.ai_message_count == 3:
                self.ready_for_more_info = True

        if role != "system":
            self.chat_display.append((content if role == "user" else None,
                                    content if role == "assistant" else None))

    def get_chat_display(self):
        return self.chat_display

    def should_show_scheduler(self):
        return self.ready_for_scheduling and not self.appointment_scheduled

    def get_messages_for_summary(self):
        return [msg for msg in self.messages if msg["role"] == "user"]

class ChatbotAI:
    def __init__(self):
        self.chat_history = ChatHistory()
        self.chat_history.initialize()

    def _get_model_response(self):
        global general_error
        try:
            response = client.chat.completions.create(
                messages=self.chat_history.messages,
                tools=th.get_tools() + my_local_tools,
                model=model,
                temperature=0.7,
                max_tokens=1024
            )
            self.chat_history.messages += th.run_tools(response)

            completion = client.chat.completions.create(
                model=model,
                messages=self.chat_history.messages,
                tools=th.get_tools() + my_local_tools,
                temperature=0.7,
                max_tokens=1024,
            )

            response_content = completion.choices[0].message.content
            return response_content
            # return response.choices[0].message.content
        except Exception as e:
            general_error = str(e)
            print(f"Error getting model response: {str(e)}")
            return "Mi dispiace, c'è stato un problema. Potresti ripetere?"

    def get_response(self, user_message, is_audio=False, final_phase = False):
        if not user_message:
            return self.chat_history.chat_display, False, False

        try:
            if final_phase:
              pass
            if is_audio:
                transcribed_text = self.transcribe_audio(user_message)
                if not transcribed_text:
                    self.chat_history.add_message("assistant", "Mi dispiace, non sono riuscito a capire l'audio. Potresti riprovare?")
                    return self.chat_history.chat_display, False, False
                user_message = transcribed_text

            self.chat_history.add_message("user", user_message)

            if self.chat_history.ready_for_more_info and not self.chat_history.ready_for_scheduling:
                if not self.chat_history.more_info_asked:
                    assistant_response = """Se ci sono altre informazioni utili che credi di dovermi dare
                                            scrivile pure, poi procederemo alla pianificazione della visita."""
                    self.chat_history.more_info_asked = True
                else:
                    self.chat_history.ready_for_scheduling = True
                    try:
                        messages_for_summary = self.chat_history.get_messages_for_summary()
                        appointment['description'] = get_summary(messages_for_summary)
                    except Exception as e:
                        print(f"Error getting summary: {str(e)}")
                        appointment['description'] = "Nessuna descrizione disponibile"

                    try:
                        suggestion, urgency = get_medical_suggestion(appointment['description'])
                        appointment['llm_suggestion'] = suggestion
                        appointment['urgency'] = urgency
                    except Exception as e:
                        print(f"Error getting medical suggestion: {str(e)}")
                        appointment['llm_suggestion'] = "Nessun suggerimento medico disponibile"
                        appointment['urgency'] = 0

                    assistant_response = ("Bene, ora procediamo con la pianificazione della visita. "
                                        "Seleziona una data e un orario tra quelli disponibili.")
            else:
                assistant_response = self._get_model_response()

                if self.chat_history.ai_message_count == 3:
                    self.chat_history.ready_for_more_info = True

            self.chat_history.add_message("assistant", assistant_response)
            return self.chat_history.chat_display, self.chat_history.should_show_scheduler(), False

        except Exception as e:
            print(f"Error in get_response: {str(e)}")
            self.chat_history.add_message("assistant", "Si è verificato un errore. Potresti riprovare?")
            return self.chat_history.chat_display, False, False

    def confirm_appointment(self, date, time):
        try:
            if not date or not time:
                return self.chat_history.get_chat_display()

            appointment["appointment_date"] = date
            appointment["appointment_time"] = time
            appointment["status"] = "confirmed"

            try:
                send_new_appointment()
            except Exception as e:
                print(f"Error sending appointment: {str(e)}")
                # Continue even if sending fails - we'll still show the confirmation

            confirmation_message = f"Perfetto, ho richiesto un appuntamento per il {date} alle {time}. Tutte le informazioni che mi hai fornito verranno inviate al tuo medico."
            self.chat_history.add_message("assistant", confirmation_message)
            # self.chat_history.add_message("system", "Se non hai già memorizzato l'indirizzo email dell'utente chiedigli di scrivertelo. Una volta ricevuto o se lo hai già inviagli una mail che contiene le informazioni riassuntive che vengono inviate al medico.")
            self.chat_history.appointment_scheduled = True
            return self.chat_history.get_chat_display()
        except Exception as e:
            print(f"Error in confirm_appointment: {str(e)}")
            error_message = "Mi dispiace, c'è stato un problema con la prenotazione. Riprova più tardi."
            self.chat_history.add_message("assistant", error_message)
            return self.chat_history.get_chat_display()

    @staticmethod
    def transcribe_audio(audio_path):
        recognizer = sr.Recognizer()
        try:
            with sr.AudioFile(audio_path) as source:
                audio_data = recognizer.record(source)
                text = recognizer.recognize_google(audio_data, language='it-IT')
                return text
        except Exception as e:
            print(f"Transcription error: {str(e)}")
            return None

def get_available_dates():
    return list(AVAILABLE_SLOTS.keys())

def get_available_times(selected_date):
    return AVAILABLE_SLOTS.get(selected_date, [])

chatbot = ChatbotAI()

with gr.Blocks(theme=gr.themes.Ocean()) as demo:
    patient_name = "Mario Rossi"
    # Header
    with gr.Row():
        with gr.Column(scale=8):
            title = gr.Markdown(
                value=f"# 👋 Buongiorno, {patient_name}\n### 🗣️ Raccontami cosa ti preoccupa, ti aiuterò a spiegarlo al tuo medico nel modo migliore",
                show_label=False
            )
        with gr.Column(scale=1):
            clear = gr.Button("🗑️", size="sm", variant="secondary")

    # Chat interface
    chatbot_interface = gr.Chatbot(
        height=400,
        bubble_full_width=False,
        show_label=False,
        container=True,
        value=[]
    )

    # Input section
    with gr.Row() as input_row:
        input_mode = gr.Radio(
            choices=["Testo", "Audio"],
            value="Testo",
            label="Modalità di input"
        )

        with gr.Column(scale=5, visible=True) as text_column:
            msg = gr.Textbox(
                placeholder="Scrivi qui il tuo messaggio...",
                lines=4,
                show_label=False
            )

        with gr.Column(scale=5, visible=False) as audio_column:
            audio = gr.Audio(
                sources=["microphone"],
                type="filepath",
                label="Registra il tuo messaggio"
            )

        with gr.Column(scale=1, min_width=100):
            submit = gr.Button("Invia")

    # Scheduler section
    with gr.Row(visible=False) as scheduler_row:
        with gr.Column(scale=2):
            date_picker = gr.Dropdown(
                choices=get_available_dates(),
                label="Seleziona una data disponibile",
                interactive=True
            )
        with gr.Column(scale=2):
            time_picker = gr.Dropdown(
                choices=[],
                label="Seleziona un orario",
                interactive=True
            )
        with gr.Column(scale=1):
            confirm_appointment = gr.Button("Conferma appuntamento", variant="primary")

    # Summary section
    with gr.Row(visible=False) as send_row:
        send_button = gr.Button("📋 Invia al Medico", variant="primary")
        summary_text = gr.Textbox(label="Riepilogo Medico", visible=False)

    def handle_audio_input(audio_data):
        if audio_data is None:
            return {
                chatbot_interface: [],
                scheduler_row: gr.update(visible=False),
                input_row: gr.update(visible=True),
                text_column: gr.update(visible=False),
                audio_column: gr.update(visible=True)
            }

        chat_history, show_scheduler, _ = chatbot.get_response(audio_data, is_audio=True)

        return {
            chatbot_interface: chat_history,
            scheduler_row: gr.update(visible=show_scheduler),
            input_row: gr.update(visible=not show_scheduler),
            text_column: gr.update(visible=False),
            audio_column: gr.update(visible=True)
        }

    def process_message(message, mode):
        if mode == "Audio":
            if audio.value:
                return handle_audio_input(audio.value)
        elif message:
            chat_history, show_scheduler, _ = chatbot.get_response(message)
            return {
                chatbot_interface: chat_history,
                scheduler_row: gr.update(visible=show_scheduler),
                input_row: gr.update(visible=not show_scheduler),
                msg: gr.update(value=""),
                text_column: gr.update(visible=mode == "Testo"),
                audio_column: gr.update(visible=mode == "Audio")
            }
        return {
            chatbot_interface: [],
            scheduler_row: gr.update(visible=False),
            input_row: gr.update(visible=True),
            msg: gr.update(value=""),
            text_column: gr.update(visible=mode == "Testo"),
            audio_column: gr.update(visible=mode == "Audio")
        }

    def toggle_input_mode(choice):
        return {
            text_column: gr.update(visible=choice == "Testo"),
            audio_column: gr.update(visible=choice == "Audio")
        }

    def update_time_slots(date):
        return gr.Dropdown(choices=get_available_times(date))

    def confirm_appointment_and_show_send(date, time):
      try:
          if not date or not time:
              return {
                  chatbot_interface: chatbot.chat_history.get_chat_display(),
                  scheduler_row: gr.update(visible=True),
                  send_row: gr.update(visible=False),
                  input_row: gr.update(visible=False)
              }

          chat_history = chatbot.confirm_appointment(date, time)
          return {
              chatbot_interface: chat_history,
              scheduler_row: gr.update(visible=False),
              send_row: gr.update(visible=True),
              input_row: gr.update(visible=False)
          }
      except Exception as e:
          print(f"Error in confirm_appointment_and_show_send: {str(e)}")
          # Return a safe fallback state
          return {
              chatbot_interface: chatbot.chat_history.get_chat_display(),
              scheduler_row: gr.update(visible=True),
              send_row: gr.update(visible=False),
              input_row: gr.update(visible=False)
          }

    def clear_chat():
        return {
            chatbot_interface: chatbot.chat_history.get_chat_display(),
            scheduler_row: gr.update(visible=False),
            send_row: gr.update(visible=False),
            input_row: gr.update(visible=True),
            msg: gr.update(value="")
        }

    # Event handlers
    input_mode.change(
        lambda choice: {
            text_column: gr.update(visible=choice == "Testo"),
            audio_column: gr.update(visible=choice == "Audio")
        },
        inputs=[input_mode],
        outputs=[text_column, audio_column]
    )

    msg.submit(
        lambda m: process_message(m, "Testo"),
        inputs=[msg],
        outputs=[chatbot_interface, scheduler_row, input_row, msg, text_column, audio_column]
    )

    submit.click(
        process_message,
        inputs=[msg, input_mode],
        outputs=[chatbot_interface, scheduler_row, input_row, msg, text_column, audio_column]
    )

    audio.stop_recording(
        handle_audio_input,
        inputs=[audio],
        outputs=[chatbot_interface, scheduler_row, input_row, text_column, audio_column]
    )

    date_picker.change(update_time_slots, inputs=[date_picker], outputs=[time_picker])
    confirm_appointment.click(
        confirm_appointment_and_show_send,
        inputs=[date_picker, time_picker],
        outputs=[chatbot_interface, scheduler_row, send_row, input_row]
    )

    clear.click(clear_chat, outputs=[chatbot_interface, scheduler_row, send_row, input_row, msg])
    # clear.click(
    #     process_message,
    #     inputs=[msg, input_mode],
    #     outputs=[chatbot_interface, scheduler_row, input_row, msg, text_column, audio_column]
    # )
    # send_button.click(
    #     process_message,
    #     inputs=[msg, input_mode],
    #     outputs=[chatbot_interface, scheduler_row, input_row, msg, text_column, audio_column]
    # )
    send_button.click(clear_chat, outputs=[chatbot_interface, scheduler_row, send_row, input_row, msg])

if __name__ == "__main__":
    demo.launch()
```
